#### Load Packages

```julia
using MatrixImpute, CSV, DataFrames, SparseArrays, Random, PyPlot, Statistics
```

#### Load Data

```julia
movLens = CSV.read("u.data", header = false, copycols = true);
```

#### Reshape to wide format

```julia
movLens = Array(sparse(movLens[!,:Column1], movLens[!,:Column2], movLens[!,:Column3]));
```

```julia
movLens = convert(Array{Union{Missing,Int64}}, movLens);
```

```julia
missidx = findall(movLens .== 0)
movLens[missidx] .= missing;
```

```julia
movLens
```

#### Investigate Scaling Behaviour

```julia
times = []
```

```julia
for k=1:5:101
    time = @elapsed begin
        out = Impute(movLens, k, method=:fastImpute, B=nothing)
    end
    push!(times, time)  
end 
```

```julia
k = [1:5:101;]
PyPlot.plot(k,times)
```

We see that the matrix completion algorithm does not scale particularly well for large values of $k$, in particular the runtime more than doubles as we increase from $k=80$ to $k=100$. However this should not be an issue in practice given the low-rank assumption. Indeed, we find that the algorithm is very fast for $k \leq 20$.

#### Convert to Missing

```julia
nmiss = findall(!ismissing, movLens)
toImp = shuffle(nmiss)[1:20000];
```

```julia
movLensTrain = deepcopy(movLens);
```

```julia
movLensTrain[toImp] .= missing;
```

#### Run MatrixImpute.jl on training matrix

```julia
MAPE_arr = [];
```

```julia
for k=1:10:200
    
    out = Impute(movLensTrain, k, method=:fastImpute, B=nothing)
    imp = out[1]*out[2]
    MAPE = mean(abs.((movLens[toImp] .- imp[toImp])./movLens[toImp]))*100
    push!(MAPE_arr, MAPE)
end;
```

```julia
PyPlot.plot([1:10:200;], MAPE_arr)
```

We see in general that as $k$ increases, the MAPE increases, peaking at roughly $42\%$ before decreasing for large $k$. This supports the low-rank assumption as we can achieve the smallest error for small values of $k$ in reasonable running times.
